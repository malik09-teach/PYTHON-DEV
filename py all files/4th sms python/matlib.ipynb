```python
import matplotlib.pyplot as plt
```

```python

x = [1, 2, 3, 4, 5]
y = [10, 25, 30, 55, 70]

plt.plot(x, y, marker='*')

plt.xlabel('x_axis')
plt.ylabel('y_axis') 
plt.title("prices and elements")

plt.show()
```

```python

ages = [22, 23, 25, 22, 45, 23, 50, 51, 22, 24, 55, 60, 62, 21, 22]

plt.hist(ages, bins=5, color='skyblue', edgecolor='black')

plt.xlabel('Age Groups')
plt.ylabel('Frequency (Count)')
plt.title('Age Distribution')

plt.show()
```
