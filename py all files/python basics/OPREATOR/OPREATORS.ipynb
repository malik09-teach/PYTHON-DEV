```python
 
# BITWISE OPREATOR 
 
# &,|,^,~,<<,>>

NUM1 =10
NUM2=4

BITAND= NUM1 & NUM2
BITOR= NUM1 | NUM2
BITXOR= NUM1 ^ NUM2
BITNOT = ~NUM1
BITWISELEFTSHIFT= NUM1 <<2
BITWISERIGHTSHIFT= NUM1 >>2


print("BITWISE AND :",BITAND)
print("BITWISE OR :",BITOR)
print("BITWISE XOR :",BITXOR)
print("BITWISE NOT :",BITNOT)
print("BITWISE LEFT SHIFT :",BITWISELEFTSHIFT)
print("BITWISE RIGHT SHIFT :",BITWISERIGHTSHIFT)

```
