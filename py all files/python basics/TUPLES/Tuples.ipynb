```python
# nested tuples 
tups=(("malik ","khan","kps"),(1,2,3,4,5,6,7,8),('a','b'))


# accessing by on indexing the main members 
print("by on indexing the main members")
print(tups[0])
print(tups[1])
print(tups[2])


# by on index accesing sub member

print ("by on index accesing sub member")
print (tups[1][1])
print (tups[1][0:])
print (tups[1][0:3])
print (tups[1][0::2])
print (tups[1][::-1])


print("by on index accesing members on the ittration")

for maj in tups:
    for sub in maj:

     print(sub,end=" ")
    print() 
```
