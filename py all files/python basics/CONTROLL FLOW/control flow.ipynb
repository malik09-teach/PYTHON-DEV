```python
    ## while loop 
print (" simple while loop :")
i=0
while i<10:
 print (i)
 i+=1

print (" break  contain loop")

i=2

while i<10:
    if i==8:
        break
    print (i)
    i+=1



print(" continue contain loop")

i=2

while i<10:
    if i==8:
        i+=1  # First increment to skip 8 and avoid infinite loop
        continue
    print(i)
    i+=1  # Second increment for normal loop progression


  
```
