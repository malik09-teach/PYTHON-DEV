```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

    @abstractmethod
    def drive(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car engine starting.")
    def drive(self):
        print("Car is driving.")

my_car = Car() # Works fine!

class Motorcycle(Vehicle):
    def start_engine(self):
        print("Motorcycle engine starting.")
    def drive(self):
        print ("the car is driving ...")
    def stopengine(self):
        print ("engine is stoped ..")    

```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

    @abstractmethod
    def drive(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car engine starting with a turn of the key.")
    def drive(self):
        print("Car is driving on four wheels.")
    def honk(self): # A method *only* in Car, not in Vehicle
        print("Beep beep!")

class Motorcycle(Vehicle):
    def start_engine(self):
        print("Motorcycle engine roaring to life.")
    def drive(self):
        print("Motorcycle is cruising on two wheels.")
    def honk(self): # A method *only* in Car, not in Vehicle
        print("Beep beep!")
    

# --- Demonstrating Polymorphism ---

# Here, 'my_transport' is a variable that 'expects' a Vehicle object.
# But we assign a Car object to it.
my_transport: Vehicle = Car() # Type hint 'Vehicle'
my_transport.start_engine() # Calls Car's start_engine()
my_transport.drive()        # Calls Car's drive()
my_transport.honk()

print("-" * 20)

# Now, 'my_transport' holds a Motorcycle object.
my_transport:Vehicle = Motorcycle() # Type hint 'Vehicle' still applies
my_transport.start_engine() # Calls Motorcycle's start_engine()
my_transport.drive()        # Calls Motorcycle's drive()
my_transport.honk()
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        """Abstract: Every vehicle must have a way to start."""
        pass

    @abstractmethod
    def drive(self):
        """Abstract: Every vehicle must have a way to drive."""
        pass

    def get_info(self): # A concrete method in the abstract base class
        return "This is a generic vehicle."

class Car(Vehicle):
    def start_engine(self): # Overridden from Vehicle
        print("Car: Engine purrs to life with a key.")
    def drive(self): # Overridden from Vehicle
        print("Car: Driving smoothly on four wheels.")
    def honk(self): # ONLY in Car, not in Vehicle
        print("Car: Beep! Beep!")

class Truck(Vehicle):
    def start_engine(self): # Overridden from Vehicle
        print("Truck: Diesel engine rumbles to life.")
    def drive(self): # Overridden from Vehicle
        print("Truck: Hauling cargo with heavy duty.")
    def unload_cargo(self): # ONLY in Truck, not in Vehicle
        print("Truck: Unloading cargo.")

# --- How to use for your desired behavior ---

# 1. Declare a variable that conceptually represents a 'Vehicle'.
#    (Using type hint 'Vehicle' for clarity, though Python doesn't enforce it strictly at runtime).
my_transport: Vehicle

my_transport = Car()

print("--- Using Car via Vehicle reference ---")
my_transport.start_engine() # Calls Car's start_engine()
my_transport.drive()        # Calls Car's drive()
print(my_transport.get_info()) # Calls Vehicle's get_info() (not overridden)

my_transport.honk()

print("\n" + "="*30 + "\n")

my_transport = Truck()

print("--- Using Truck via Vehicle reference ---")
my_transport.start_engine() 
my_transport.drive()        
print(my_transport.get_info()) 
my_transport.unload_cargo()
```
