```python
# Example using functools.singledispatch (built-in in Python 3)
from functools import singledispatch

@singledispatch
def process_item(item):
    """Default behavior for processing an item."""
    return f"Unknown item type: {type(item).__name__}"

@process_item.register(int)
def _(item: int):
    return f"Processing integer: {item * 10}"

@process_item.register(str)
def _(item: str):
    return f"Processing string: {item.lower()}"

# Note: singledispatch typically applies to functions, not methods within a class directly
# for methods, you'd combine it with a class method or similar approach, or use multipledispatch

print(process_item(10))      # Output: Processing integer: 100
print(process_item("HELLO")) # Output: Processing string: hello
print(process_item(3.14))    # Output: Unknown item type: float
```
