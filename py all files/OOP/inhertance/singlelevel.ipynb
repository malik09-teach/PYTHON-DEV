```python
#3 single level 
class A:
    def __init__(self, name ,age ,roll,dept):
        self.name=name
        self.age=age
        self.roll=roll
        self.dept=dept

    def disp_info(self):
         print("from class A")
         print(self.name)
         print(self.age)
         print( self.roll)
         print(self.dept)

```

```python
## syntax of inherit the class  

class B(A):
    def __init__(self, name ,age ,roll,dept,subject):
           super().__init__(name ,age ,roll,dept)
           self.subjects=subject

    def disp_info(self):
         print("from class B")
         print(self.name)
         print(self.age)
         print( self.roll)
         print(self.dept)
         print("subjects:",self.subjects)

```

```python
## objects 
st1=A("malik",23,2290,"cs")
s1=B("malik",23,2290,"cs",["enlish","pf","ict"])

st1.disp_info()
s1.disp_info()
```
