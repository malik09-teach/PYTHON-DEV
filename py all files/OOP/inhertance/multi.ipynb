```python
class Company:
    def __init__ (self,product,name,price,item):
        self.product=product
        self.name=name
        self.price=price
        self.item=item 
    def  product_detail(self):
        print(f"name:{self.name}\nprice/suit:{self.price}\nproduct:{self.product}\nitems avilible:{self.item}")
        print("amount taken from resaler:",self.purchasecalculation()) 
        
    def purchasecalculation(self):
        return self.item*self.price
    
```

```python
class Resaler(Company):

         
    
    def __init__(self,product,name,price,item,cust):
         super().__init__(product,name,price,item)
         self.cust=cust     
         self.priceown=price+550      

    def product_detail(self):
         print(f"name:{self.name}\nprice/suit:{self.priceown}\nproduct:{self.product}\nitems avilible:{self.item}\ntotal amount genarate:{self.purchasecalculation()}\ncustormers:{self.__customeravilible()}")
    def __customeravilible(self):
         return self.cust
    def purchasecalculation(self):
         return self.item*self.priceown

         
    
   
          
```

```python
class Customer(Resaler):

    def __init__(self,name,item):
        self.item=item
        self.name=name
    

    def purchasecalculation(self):
        return 1250*self.item

    def detailofshoping(self):
        print(f"name:{self.name}\nitems:{self.item}\ntotal amount:{self.purchasecalculation()}")
         
```
