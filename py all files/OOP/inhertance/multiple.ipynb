```python
#calls 
class Music:
    def __init__(self,name,duration,artist):
        self.name=name
        self.duration=duration
        self.artist=artist
    
    def song_info(self):
     print("info..")
```

```python
class Publish:   
    def __init__(self,guest,location):       
        self.guest=guest
        self.location=location

    def infosofpub (self):
        print("sub info...")    
```

```python
class Concert (Music,Publish):
    __tickettot=150
    __pertekprice=3500
 
    def __init__(self,guest,location,name,duration,artist,tickets):
        Music.__init__(self,name,duration,artist)
        Publish.__init__(self,guest,location)
        self.tickets=tickets


    def amount_from_tickets(self):
         return (self.__tickettot-self.tickets)*self.__pertekprice 
    
    def song_info(self):
     print("info..of song..")
     print("song name:",self.name)
     print("artist:",self.artist)
     print("duratiion:",self.duration,"min")
    
    def infosofpub(self):
        print("  info of concert...") 
        print("location:",self.location) 
        print ("the private guest is:",self.guest[::])

    def overalldetail(self):
        self.song_info()
        self.infosofpub()
        print("totalamount:",amount_from_tickets())
```
