```python
class A:
      def __init__(self,name):
          self.name=name 
      def method1(self):
          print ("name of method..m1 :",self.name)  
```

```python
class B:
    def __init__(self,name2):
          self.name2=name2 
    def method(self):
          print ("name of method m..:",self.name2)  
```

```python
class mid(A,B):
    def __init__(self,name,name2,name3):
        A.__init__(self,name)
        B.__init__(self,name2)
        self.name3=name3
# if we dont override here so the method will direct called from the a,b class 
    def method1(self):
      print("calling of mid method mehtod1...")
    def method(self):
        print("calling of mid method mehtod...")

    def midmethod(self):
    
        print("this is mid level method printing...",self.name3)


```
