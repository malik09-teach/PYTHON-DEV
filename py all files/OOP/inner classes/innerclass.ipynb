```python
class outer :
    def __init__(self,num1,num2):
        self.num1=num1
        self.num2=num2
    
    def calculation(self):
        return self.num1**self.num2    
    
    class inner:
        def __init__(self,name,brand,price):
            self.name=name
            self.brand=brand
            self.price=price  
        
        def branding(self):
           print (f"name :{self.name} \nbrand :{self.brand} \nprice:{self.price}")   
       
   
## object calling 
obj1=outer(5,6)
ob2=outer.inner("white","crolla",3200000)

print("power:",obj1.calculation())
ob2.branding()
print(obj1)

```

```python
## let make a program
class laptop:
    def __init__ (self,name,cpu):
        self.name=name 
        self.cpu=cpu

    def lapinfo(self):
        print (f"lap_name:{self.name} \ncpu:{self.cpu}\n****************")

    class hideinfo:
        def __init__(self,macAdd,network,ip):
            self.macAdd=macAdd
            self.network= network
            self.ip=ip


        def hideinfolap(self):   
            print(f"mac add:{self.macAdd}  \nnetwork: {self.network } \nip: {self.ip}")

object1=laptop("ryzen","R9095x")
obj2=laptop.hideinfo("Xf94900900","4G+","190.265.1.2")

object1.lapinfo()
obj2.hideinfolap()

```
