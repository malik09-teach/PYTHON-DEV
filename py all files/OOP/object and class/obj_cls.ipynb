```python
class MyMathUtils:
    # A regular class attribute
    PI = 3.14159

    # 1. Instance Method
    # It receives 'self' and can access instance-specific data.
    def get_instance_info(self):
        return f"This is an instance method of {self.__class__.__name__}."

    # 2. Class Method
    # It receives 'cls' (the class itself) and can access class-level data.
    @classmethod
    def get_class_info(cls):
        return f"This is a class method of {cls.__name__}."

    # 3. Static Method
    # It receives no implicit first argument ('self' or 'cls').
    @staticmethod
    def calculate_circle_area(radius):
        # We can still access class attributes if we explicitly use the class name
        # but the method itself is independent of the instance or class.
        return MyMathUtils.PI * (radius ** 2)

# --- How to use the methods ---

# Create an instance of the class
my_object = MyMathUtils()

# Call the instance method on the object
print(my_object.get_instance_info())

# Call the class method on the class
print(MyMathUtils.get_class_info())

# Call the static method on the class (common practice)
area_from_class = MyMathUtils.calculate_circle_area(5)
print(f"Area of a circle with radius 5 (called on class): {area_from_class}")

# Call the static method on the instance (also works, but not common)
area_from_object = my_object.calculate_circle_area(10)
print(f"Area of a circle with radius 10 (called on instance): {area_from_object}")
```
